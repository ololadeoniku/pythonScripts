```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("../trace_log_ta_ch1_A.csv", skiprows = 3)
fileobj = file.copy()
# fileobj.columns.values[0], fileobj.columns.values[1] = 'Time', 'ACBias_V'
# fileobj.set_index('Time', inplace = True)
# fileobj.sort_index(inplace=True)
fileobj.columns.values
# fileobj.
cols = np.array(['Time', 'ACBias_V', 'BiasMatchLoad_mV', 'BiasMatchTune_mV', 'ACBiasForward_W', 'ACBiasReflected_W', 'Ar_Backside_sccm',
 'Ar_Flow_sccm', 'BacksidePress_torr', 'CenterTapBias_V', 'CryoTemp_degK', 'DCCurrent_A', 'DCPower_W', 'TargetVoltage',
 'DCArc_Acc_Count', 'DCArc_Acc_Micro_Count', 'DCArc', 'DCArc_Micro', 'ESC_Current_mA', 'ESC_Voltage_V','Pedestal_Temp_degC',
 'HtrLift_Step_step', 'HX_Temp_degC', 'IG_Pressure_torr', 'Lift_Position', 'Magnet_Pos', 'N2_Flow_sccm', 'CHPressure_torr',
 'Step Number', 'TargetLife_kWh'])
```

```python
columns = np.array(['Time', 'Step Number','TargetLife_kWh', 'TargetVoltage', 'DCPower_W', 'ACBiasForward_W', 'ACBias_V', 
        'BiasMatchLoad_mV', 'BiasMatchTune_mV', 'DCCurrent_A', 'ACBiasReflected_W', 'CenterTapBias_V', 'ESC_Voltage_V', 
        'ESC_Current_mA','BacksidePress_torr', 'CHPressure_torr', 'Ar_Flow_sccm', 'N2_Flow_sccm', 'Ar_Backside_sccm', 
        'IG_Pressure_torr', 'Magnet_Pos', 'Lift_Position', 'HtrLift_Step_step', 'CryoTemp_degK', 'Pedestal_Temp_degC', 
        'HX_Temp_degC', 'DCArc_Acc_Count', 'DCArc_Acc_Micro_Count',
       'DCArc', 'DCArc_Micro'])
```

```python
fileobj.columns = cols
```

```python
fileobj = fileobj[['Time', 'Step Number','TargetLife_kWh', 'TargetVoltage', 'DCPower_W', 'ACBiasForward_W', 'ACBias_V', 
        'BiasMatchLoad_mV', 'BiasMatchTune_mV', 'DCCurrent_A', 'ACBiasReflected_W', 'CenterTapBias_V', 'ESC_Voltage_V', 
        'ESC_Current_mA','BacksidePress_torr', 'CHPressure_torr', 'Ar_Flow_sccm', 'N2_Flow_sccm', 'Ar_Backside_sccm', 
        'IG_Pressure_torr', 'Magnet_Pos', 'Lift_Position', 'HtrLift_Step_step', 'CryoTemp_degK', 'Pedestal_Temp_degC', 
        'HX_Temp_degC', 'DCArc_Acc_Count', 'DCArc_Acc_Micro_Count',
       'DCArc', 'DCArc_Micro']]
fileobj.head()
```

```python
fileobj['Time'] = pd.to_timedelta(fileobj['Time'], errors = 'coerce')
fileobj['Time'] = fileobj['Time'].dt.total_seconds() - fileobj['Time'].dt.total_seconds().iloc[0]
fileobj.head()
for column in columns:
    fileobj[column] = pd.to_numeric(fileobj[column], errors = 'coerce')
fileobj.isnull()
fileobj.head()
# for column in columns:
#     print(type(column))
```

```python
fileobj = fileobj[(fileobj['Step Number'] > 0) & (fileobj['Step Number'] != 'NaN') & (fileobj['Step Number'] < 21)]
fileobj.groupby('Step Number').agg({'TargetVoltage': ['min', 'max', 'count'], 'TargetLife_kWh':['mean']}).sort_index(ascending=True)
fileobj[['Time','Step Number', 'TargetVoltage', 'TargetLife_kWh']].head(460)
# fileobjNoNaN.groupby('Step Number').agg({'DCSrc_Voltage_Get (V)': ['min', 'max', 'count']}).sort_index(ascending=
#                                                 True).unstack('Target Life', fill_value = 0)
```
